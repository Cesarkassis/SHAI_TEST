```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer



df = pd.read_csv('/content/Salaries.csv')

# Drop unnecessary columns
df = df[['Id', 'EmployeeName', 'JobTitle', 'BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay', 'TotalPayBenefits', 'Year', 'Agency']]

# Identify and handle missing values
numerical_columns = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']

# Impute missing values in numerical columns
numerical_imputer = SimpleImputer(strategy='mean')
df.loc[:, numerical_columns] = numerical_imputer.fit_transform(df[numerical_columns])

# Separate data into features and target
features = df[df['TotalPay'] != 0][numerical_columns]
target = df[df['TotalPay'] != 0]['TotalPay']

# Train a linear_regression model
model = LinearRegression()
model.fit(features, target)

# Predict 'TotalPay' for rows with 'TotalPay' equal to zero
missing_data = df[df['TotalPay'] == 0][numerical_columns]
predicted_totalpay = model.predict(missing_data)

# Fill in missing 'TotalPay' values with predictions
df.loc[df['TotalPay'] == 0, 'TotalPay'] = predicted_totalpay


print(df)
```

```python
# Task 4: Create histograms to visualize the distribution of salaries and Use pie charts to represent the proportion of employees in different departments

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv('/content/Salaries.csv')

# Drop unnecessary columns
df = df[['Id', 'EmployeeName', 'JobTitle', 'BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay', 'TotalPayBenefits', 'Year', 'Agency']]

# Identify and handle missing values
numerical_columns = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']

# Impute missing values in numerical columns
numerical_imputer = SimpleImputer(strategy='mean')
df.loc[:, numerical_columns] = numerical_imputer.fit_transform(df[numerical_columns])

# Separate data into features and target
features = df[df['TotalPay'] != 0][numerical_columns]
target = df[df['TotalPay'] != 0]['TotalPay']

# Train a linear_regression model
model = LinearRegression()
model.fit(features, target)

# Predict 'TotalPay' for rows with 'TotalPay' equal to zero
missing_data = df[df['TotalPay'] == 0][numerical_columns]
predicted_totalpay = model.predict(missing_data)

# Fill in missing 'TotalPay' values with predictions
df.loc[df['TotalPay'] == 0, 'TotalPay'] = predicted_totalpay

# Basic Data Visualization
# Histogram of 'TotalPay'
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'], bins=30, kde=True)
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()

# Bar chart for the proportion of employees in different departments
plt.figure(figsize=(12, 8))
df['JobTitle'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Job Titles with Employee Count')
plt.xlabel('Job Title')
plt.ylabel('Number of Employees')
plt.show()

# Pie chart for the proportion of employees in different departments
plt.figure(figsize=(12, 12))
df['JobTitle'].value_counts().head(10).plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportion of Employees in Top 10 Job Titles')
plt.show()
```

```python
# Task 5: Grouped Analysis

grouped_by_agency = df.groupby('Agency')['TotalPay'].describe()

# Displaying summary statistics for each agency
print("Summary Statistics for TotalPay Grouped by Agency:")
print(grouped_by_agency)

# Comparing average salaries across different agencies
average_salary_by_agency = df.groupby('Agency')['TotalPay'].mean()

plt.figure(figsize=(12, 6))
average_salary_by_agency.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Average TotalPay Across Different Agencies')
plt.xlabel('Agency')
plt.ylabel('Average TotalPay')
plt.show()
```

```python
# Task 6: Simple Correlation Analysis

# Checking correlation between 'TotalPay' and 'BasePay'
correlation = df['TotalPay'].corr(df['BasePay'])

print(f"Correlation between TotalPay and BasePay: {correlation}")

plt.figure(figsize=(10, 6))
plt.scatter(df['BasePay'], df['TotalPay'], alpha=0.5, color='blue')
plt.title('Scatter Plot: TotalPay vs BasePay')
plt.xlabel('BasePay')
plt.ylabel('TotalPay')
plt.show()
```
